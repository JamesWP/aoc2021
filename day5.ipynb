```python
import matplotlib.pyplot as plt
import numpy
import seaborn
import math
```

```python
lines = []
with open("input5.txt", "r") as data:
    for line in data.readlines():
        coords = [[int(c) for c in coord.split(",")] for coord in line.strip().split("->")]
        lines.append(coords)
lines = numpy.reshape(numpy.array(lines), (-1, 4))
```

```python
horiz_or_vert = numpy.logical_or(lines[:,0] == lines[:,2], lines[:,1] == lines[:,3])
```

```python
shape = numpy.max(lines[:, (0,2)])+1, numpy.max(lines[:,(1,3)])+1
space = numpy.zeros(shape, dtype=int)
```

```python
xs = numpy.arange(space.shape[0])
ys = numpy.arange(space.shape[1])

def draw_line(canvas, coords):
    x1,y1,x2,y2 = coords
    
    start = numpy.array([x1,y1])
    end = numpy.array([x2,y2])

    length = max(abs(x2-x1), abs(y2-y1))+1
    print(coords, length)
    coord = start
    for step in range(length):
        canvas[coord[0], coord[1]] += 1
        if x2 > x1:
            coord[0] +=1
        if x2 < x1:
            coord[0] -=1
        if y2 > y1:
            coord[1] +=1
        if y2 < y1:
            coord[1] -=1
```

```python
for line in lines:
    draw_line(space, line)
```
